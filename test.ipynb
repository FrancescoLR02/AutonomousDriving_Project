```python
import time
import gymnasium
import torch
import highway_env


# 1. Force CPU
device = torch.device('cpu')
print(f"Testing on: {device}")

# 2. No Rendering
env = gymnasium.make("highway-fast-v0", render_mode=None)
obs, _ = env.reset()

start_time = time.time()
steps = 1000

for _ in range(steps):
    # Simulate the overhead of your Agent
    obs_tensor = torch.as_tensor(obs, dtype=torch.float32, device=device)
    action = env.action_space.sample() # Random action is instant
    obs, reward, done, _, _ = env.step(action)
    if done:
        obs, _ = env.reset()

end_time = time.time()
duration = end_time - start_time
fps = steps / duration

print(f"Speed: {fps:.2f} steps/second")
print(f"Predicted time for 20k steps: {(20000/fps)/60:.2f} minutes")
```
